```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.optimize import curve_fit
```

```python
n = 15
xi = np.linspace(-2,4, n)
A = np.array([ xi, np.ones(n)])
yi = 2 * xi + 1 + np.random.normal(0, 1, size=n)
```

```python
a, b, _, _, _ = stats.linregress(xi,yi)
```

```python
yi_predict = a * xi + b
ax = plt.plot(xi, yi_predict, 'r-', xi, yi,'o')
```

```python
def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))
yi = logistic(xi, 4, 2, 1) + np.random.normal(0, 0.35, size=n)
popt, pcov = curve_fit(logistic, xi, yi)
yi_predict = logistic(xi, *popt)
ax = plt.plot(xi, yi_predict, 'r-', xi, yi,'o')
```
